```python
import re
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
f=open("snail1.txt","r",encoding="utf-8")
```

```python
data=f.read()
```

```python
data
```

```python
pattern=r"\d{2}/\d{2}/\d{4}, \d{2}:\d{2}\s?[ap]m - "
```

```python
messages = re.split(pattern,data)
messages
```

```python
date_time = re.findall(pattern,data)
date_time
```

```python
data_d={"messages":messages[1:],"date":date_time}
```

```python
len(messages)
```

```python
len(date_time)
```

```python
df=pd.DataFrame(data_d)
```

```python
df.head()
```

```python
pattern=r".*?:\s*"
```

```python
df["user"]=df["messages"].apply(lambda x:re.search(pattern,x).group() if re.search(pattern,x) else "group notification" )
```

```python
df["user"]=df["user"].apply(lambda x:x.replace(":",""))
```

```python
df.head()
```

```python
df["messages"]=df["messages"].apply(lambda x:re.split(pattern,x))
```

```python
df.head()
```

```python
df["messages"]=df["messages"].apply(lambda x:x[0] if x[0] else x[1])
```

```python
df["messages"]=df["messages"].apply(lambda x:x.replace("\n",""))
```

```python
df.head(10)
```

```python

df["message_Date"]=pd.to_datetime(df["date"],format="%m/%d/%y, %H:%M - ")
```

```python
df=df.drop("date",axis=1)
```

```python
df["hour"]=df.message_Date.dt.hour
df["minute"]=df.message_Date.dt.minute
df["second"]=df.message_Date.dt.second
df["day"]=df.message_Date.dt.day_name()
df["month"]=df.message_Date.dt.month_name()
df["year"]=df.message_Date.dt.year
```

```python
def histogram(typ="type",name=name,number=[]):
    h=pd.DataFrame({"name":name,f"number of {typ}": number})
   
    plt.title(f"{typ}")
    plt.barh(h["name"],h[f"number of {typ}"],color="purple")
    
    plt.ylabel("Words")
    plt.xlabel(f"Number of {typ}")
    for index, value in enumerate(h[f"number of {typ}"]):
        plt.text(value, index,
                 str(value))
    plt.show()
```
